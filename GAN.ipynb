```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class Generator(nn.Module):
    def __init__(self,z_dim,  output_dim = 1):
        super(Generator, self).__init__()
   
        self.model = nn.Sequential(

            nn.Linear(z_dim, 32),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Linear(32, 64),
            #nn.BatchNorm1d(64, 0.8),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Linear(64, output_dim)
        )

    def forward(self, z):
        return self.model(z)
    

    
class Discriminator(nn.Module):
    def __init__(self, input_dim):
        super(Discriminator, self).__init__()
        
        self.model = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.LeakyReLU(0.2, inplace = True),
            
            nn.Linear(64, 32),
            nn.LeakyReLU(0.2, inplace = True),

            nn.Linear(32, 1)
        )

    def forward(self, xy):
        return self.model(xy)
```

```python
from sklearn.datasets import make_moons


import matplotlib.pyplot as plt

import torch
import torch.nn as nn



latent_shape = 10

def sample_moons(n_samples=10000):
    X, _ = make_moons(n_samples=n_samples, noise=0.1)
    return torch.tensor(X, dtype=torch.float32)




#Initializing Models and Optimizers
G = Generator(z_dim=latent_shape, output_dim=2)
D = Discriminator(input_dim=2)
criterion = torch.nn.BCEWithLogitsLoss()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
G.to(device)
D.to(device)
criterion.to(device)

def weights_init(m):
    if isinstance(m, nn.Linear):
        nn.init.normal_(m.weight, mean=0.0, std=0.02)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0)

G.apply(weights_init)
D.apply(weights_init)


optimizer_G = torch.optim.Adam(G.parameters(), lr=2e-4, betas=(0.5, 0.999))
optimizer_D = torch.optim.Adam(D.parameters(), lr=2e-4, betas=(0.5, 0.999))


#########
#TRAINING
#########
no_epochs = 40000
batch_size = 1024

real_data = sample_moons(batch_size).to(device)

for epoch in range(no_epochs):
        
    real_labels = torch.ones(batch_size, 1, device = device)
    fake_labels = torch.zeros(batch_size, 1, device = device)
        
    #Training Generator
    z = torch.randn(batch_size, latent_shape, device = device)
    gen_samples = G(z)
    D_output = D(gen_samples)

    G_loss = criterion(D_output, real_labels)

    optimizer_G.zero_grad()
    G_loss.backward()
    optimizer_G.step()

        
    #Training Discriminator
    z = torch.randn(batch_size, latent_shape, device = device)
    fake_samples = G(z).detach()
    D_real = D(real_data)
    D_fake = D(fake_samples)

    loss_real = criterion(D_real, real_labels)
    loss_fake = criterion(D_fake, fake_labels)

    D_loss = (loss_fake + loss_real) / 2

    optimizer_D.zero_grad()
    D_loss.backward()
    optimizer_D.step()

        

    if epoch % 2000 == 0:
        print(f"Epoch {epoch} | G Loss: {G_loss.item():.4f} | D Loss: {D_loss.item():.4f}")
        
        print("Sample fake:", gen_samples[:5].detach().cpu().numpy())

        fake_plot = gen_samples.detach().cpu().numpy()
        real_plot = real_data.detach().cpu().numpy()
        
        plt.figure(figsize=(6, 6))
        plt.scatter(real_plot[:, 0], real_plot[:, 1], s=5, label='Real', alpha=0.5, c='royalblue')
        plt.scatter(fake_plot[:, 0], fake_plot[:, 1], s=5, label='Fake', alpha=0.5, c='orangered')
        plt.title(f"Epoch: {epoch}")
        plt.legend()
        plt.xlim(-2, 3)
        plt.ylim(-1.5, 2)
        plt.savefig(f"moon_epoch_{epoch}.png")
        plt.show()
        plt.close()

    
    
        
        
      
```
